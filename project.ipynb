```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df=pd.read_csv("/workspaces/data-set-1/Zomato data .csv")

# Converting datatype of "rate" column to float and removing denominator :
def handleRate(value):
    value =str(value).split('/')
    value=value[0];
    return float(value)


df['rate']=df['rate'].apply(handleRate)
print(df.head)

df.info

# Type of restaurant
df=pd.read_csv("/workspaces/data-set-1/Zomato data .csv")
sns.countplot(x=df['listed_in(type)'])
plt.xlabel("Type of restaurant",c="red")
plt.show()




# Line graph depicting relation between votes and type of resaurant :
dataframe=pd.read_csv('/workspaces/data-set-1/Zomato data .csv')
grouped_data=dataframe.groupby('listed_in(type)')['votes'].sum()
result=pd.DataFrame({'votes':grouped_data})
plt.plot(result,c="blue",marker="o")
plt.xlabel("Type of restaurant",c="green",size=25)
plt.ylabel("votes",c="green",size=25)
plt.show()
plt.close



# Ratings recieved by majority of restaurants :
dataframe=pd.read_csv('/workspaces/data-set-1/Zomato data .csv')
plt.hist(dataframe['rate'],bins=5)
plt.title("Ratings Distribution")
plt.show()

# Average Spending on  each order
dataframe=pd.read_csv("/workspaces/data-set-1/Zomato data .csv")
couple_data=dataframe['approx_cost(for two people)']
sns.countplot(x=couple_data)


# which mode of order receive higher ratings online/offline orders :
plt.figure(figsize=(6,6))
sns.boxplot(x='online_order',y='rate',data=dataframe)


# What type of resaturants receives more offline orders :
dataframe=pd.read_csv("/workspaces/data-set-1/Zomato data .csv")
pivot_table = dataframe.pivot_table(index='listed_in(type)', columns='online_order', aggfunc='count')
sns.heatmap(pivot_table,annot=True,cmap="YlGn" , fmt='d')
plt.title("Heatmap")
plt.xlabel("online_order")
plt.ylabel("Listed In (Type)")
plt.show()
















```
